## Extract

```python
import pandas as pd 
from datetime import datetime

df = pd.read_csv('D:\DIO\Bootcamp_Santander\contas_a_pagar.csv', dayfirst=True, parse_dates=['DT VENCIMENTO'], date_parser=lambda x: datetime.strptime(x, '%d/%m/%Y'))
df.info()
```

```python
df
```

## Transform

```python
valor_sub = "Nao pago"
df = df.fillna(valor_sub)
df
```

```python


def pagoounão(lin):
    data_da_coluna = pd.to_datetime(lin['DT VENCIMENTO'])
    data_de_hoje = datetime.today().date()
    data_da_coluna_date = data_da_coluna.date()

    if 'Nao pago' in lin['DT PAGAMENTO'] and data_da_coluna_date < data_de_hoje:
        return 'Vencido'
    elif 'Nao pago' in lin['DT PAGAMENTO'] and data_da_coluna_date > data_de_hoje:
        return 'Pendente'
    else:
        return 'Pago'


df['STATUS'] = df.apply(pagoounão, axis=1)
df
```

```python
def tempopend(lin):
    data_da_coluna = pd.to_datetime(lin['DT VENCIMENTO'])
    data_de_hoje = datetime.today().date()
    data_da_coluna_date = data_da_coluna.date()

    if 'Nao pago' in lin['DT PAGAMENTO'] and 'Vencido' in lin['STATUS']:
        return abs((data_de_hoje-data_da_coluna_date).days)
    else:
        return 'N/A'


df['DIAS VENCIDO'] = df.apply(tempopend, axis=1)
df
```
