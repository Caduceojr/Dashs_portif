```python
novacols_dict ={
    'timestamp':'tempo',
    'cnt':'contagem',
    't1':'temp_real',
    't2':'temp_sensação',
    'hum':'humidade_porcentagem',
    'wind_speed':'vel_vento',
    'weather_code':'clima',
    'is_holiday':'feriado_bol',
    'is_weekend':'fds_bol',
    'season':'estação'
}

bikes.rename(novacols_dict, axis=1, inplace=True)
```

```python
bikes
```

```python
estac_dict = {
    '0.0':'primavera',
    '1.0':'verao',
    '2.0':'outono',
    '3.0':'inverno'
}


clima_dict = {
    '1.0':'limpo',
    '2.0':'nuvens dispersas',
    '3.0':'muitas nuvens',
    '4.0':'nublado',
    '7.0':'chuva',
    '10.0':'tempestade',
    '26.0':'neve'
}

bikes.estação = bikes.estação.astype('str')
bikes.estação = bikes.estação.map(estac_dict)

bikes.clima = bikes.clima.astype('str')
bikes.clima = bikes.clima.map(clima_dict)
```
